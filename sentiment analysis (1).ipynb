```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
input=pd.read_csv('D:\\indianbank_twitterdata\\scraped_tweets_v1.csv')
```

```python
input.rename(columns={'Unnamed: 0':'Obs'},inplace=True)
```

```python
input
```

```python
input=input.loc[input['username']!='MyIndianBank']#removing tweet of offical id myindianbank
```

```python
def find_key(list,key,substr):
    for i in list:
        if str(i).upper()==str(key).upper() or str(i[-3:]).upper()==str(substr).upper(): 
            c=1
            break
        else:
            c=0
        return c

```

```python
input.columns
```

```python
input.sort_values(["username",'location','created_dt',"cleaned_text"],inplace=True)
```

```python
input.drop_duplicates(["username",'location',"cleaned_text"],keep=False,inplace=True)
```

```python
input['hashtags'].unique()
```

```python
input=input.loc[~input['hashtags'].str.contains("ParalympicsTokyo|'JustTeacherThings'|'ContestAlert'", flags=re.IGNORECASE)]#removing paraolampics data
```

```python
input=input.loc[~input['hashtags'].str.contains("\w+Day", flags=re.IGNORECASE,regex=True)]#removing paraolampics data
```

```python
input
```

```python
senti_analyzer = SentimentIntensityAnalyzer()
input.columns
```

```python
final={'obs':[],
    'neg': [],
 'neu':[],
 'pos': [],
 'compound': [],
 'comment': [],
 'review':[]}
for i,sen in zip(input['Obs'],input['cleaned_text']):
    out=senti_analyzer.polarity_scores(sen)
    out.update({'comment':sen})
    out.update({'obs':i})
    if out['compound']>0.05:
        out['review']='Positive'
    elif out['compound']<-0.05:
        out['review']='negtive'
    else:
        out['review']='Nutral'
    for key in out.keys():
        final[key].append(out[key])
    
    
```

```python
final

```

```python
final_df=pd.DataFrame.from_dict(final)

final_df
```

```python
out=pd.merge(left=input,right=final_df,how='left',left_on='Obs',right_on='obs')
```
